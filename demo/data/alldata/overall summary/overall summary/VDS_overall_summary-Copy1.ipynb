```python
def article1_2(left_data, right_data, option1):
    if(option1 == 0):
        left_right = pd.merge(left_data, right_data,  how='outer', on='term')
        left_right['ratio_x'].fillna(0, inplace=True)
        left_right['ratio_y'].fillna(0, inplace=True)
        left_right_perc = left_right.filter(items=['term', 'ratio_x', 'ratio_y'])
        left_right_perc.rename(columns={'term':'term', 'ratio_x':'a1_perc', 'ratio_y':'a2_perc'}, inplace=True)
        min_value = left_right_perc[['a1_perc','a2_perc']].min().min()
        max_value = left_right_perc[['a1_perc','a2_perc']].max().max()
        bin_value = (max_value-min_value)/10
        left_right_perc['a1_bin'] = np.ceil(left_right_perc['a1_perc'] / bin_value)
        left_right_perc['a2_bin'] = np.ceil(left_right_perc['a2_perc'] / bin_value)
    else:
        left_right = pd.merge(left_data, right_data,  how='outer', on='tags_id')
        left_right['label_x'].fillna(left_right['label_y'], inplace=True)
        left_right['pct_x'].fillna(0, inplace=True)
        left_right['pct_y'].fillna(0, inplace=True)
        left_right_perc = left_right.filter(items=['label_x', 'pct_x', 'pct_y'])
        left_right_perc.rename(columns={'label_x':'term', 'pct_x':'a1_perc', 'pct_y':'a2_perc'}, inplace=True)
        min_value = left_right_perc[['a1_perc','a2_perc']].min().min()
        max_value = left_right_perc[['a1_perc','a2_perc']].max().max()
        bin_value = (max_value-min_value)/10
        left_right_perc['a1_bin'] = np.ceil(left_right_perc['a1_perc'] / bin_value)
        left_right_perc['a2_bin'] = np.ceil(left_right_perc['a2_perc'] / bin_value)
    dist_left_right_perc = left_right_perc[['term', 'a1_bin', 'a1_perc', 'a2_bin', 'a2_perc']]
    #dist_left_right_perc.to_csv(word_dist_file.iloc[0], index=False)
    dist_left_right_heatmap = pd.DataFrame({'count' : dist_left_right_perc.groupby(['a1_bin', 'a2_bin'])['term'].count()}).reset_index()
    return dist_left_right_perc, dist_left_right_heatmap
    #dist_left_right_heatmap.to_csv(heatmap_file.iloc[0], index=False)
```

```python
def get_type(case):
    switch = ['unigram', 'bigram', 'trigram', 'people', 'organization', 'theme']
    return switch[case]
```

```python
def top5terms(group, data, check, output):
    global overall_top5
    if(check==True):
        data.rename(columns={'term':'label', 'ratio':'perc'}, inplace=True)
    else:
        data.rename(columns={'pct':'perc'}, inplace=True)
    mean_value = data['perc'].mean()
    data[['label', 'perc']].to_csv(output.iloc[0], index=False)
    top5_data = data.nlargest(5, 'perc')[['label', 'perc']]
    top5_data['start'] = range(0, 5)
    top5_data['end'] = top5_data['start'] + 1 
    top5_data['value'] = top5_data['perc']/mean_value
    top5_data['chr'] = group
    overall_top5 = overall_top5.append(top5_data[['chr','label','start', 'end', 'value']])
```
