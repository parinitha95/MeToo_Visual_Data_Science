```python
import pandas as pd
import numpy as np
```

```python
input_files = pd.read_csv("C:\\Users\\e237513\\Desktop\\School work\\vds\\input_file_names.csv", sep=',')
output_file_names = pd.read_csv("C:\\Users\\e237513\\Desktop\\School work\\vds\\final_file_names.csv", sep=',')
location = "C:\\Users\\e237513\\Desktop\\School work\\vds\\data\\MeToo data for data visualization project\\"
```

```python
a=output_file_names
for index, row in input_files.iterrows():
    a=pd.DataFrame(output_file_names.loc[output_file_names['article_name']==row[0]])
    for i in range(1,12,2):
        heatmap_file = a.iloc[:,i]
        word_dist_file = a.iloc[:,i+1]
        left_data = pd.read_csv(location+row[i], sep=",")
        right_data=pd.read_csv(location+row[i+1], sep=",")
        if(i < 6):
            left_right = pd.merge(left_data, right_data,  how='outer', on='term')
            left_right['ratio_x'].fillna(0, inplace=True)
            left_right['ratio_y'].fillna(0, inplace=True)
            left_right_perc = left_right.filter(items=['term', 'ratio_x', 'ratio_y'])
            left_right_perc.rename(columns={'term':'term', 'ratio_x':'left_perc', 'ratio_y':'right_perc'}, inplace=True)
            min_value = left_right_perc[['left_perc','right_perc']].min().min()
            max_value = left_right_perc[['left_perc','right_perc']].max().max()
            bin_value = (max_value-min_value)/10
            left_right_perc['left_bin'] = np.ceil(left_right_perc['left_perc'] / bin_value)
            left_right_perc['right_bin'] = np.ceil(left_right_perc['right_perc'] / bin_value)
        else:
            left_right = pd.merge(left_data, right_data,  how='outer', on='tags_id')
            left_right['label_x'].fillna(left_right['label_y'], inplace=True)
            left_right['pct_x'].fillna(0, inplace=True)
            left_right['pct_y'].fillna(0, inplace=True)
            left_right_perc = left_right.filter(items=['label_x', 'pct_x', 'pct_y'])
            left_right_perc.rename(columns={'label_x':'term', 'pct_x':'left_perc', 'pct_y':'right_perc'}, inplace=True)
            min_value = left_right_perc[['left_perc','right_perc']].min().min()
            max_value = left_right_perc[['left_perc','right_perc']].max().max()
            bin_value = (max_value-min_value)/10
            left_right_perc['left_bin'] = np.ceil(left_right_perc['left_perc'] / bin_value)
            left_right_perc['right_bin'] = np.ceil(left_right_perc['right_perc'] / bin_value)
        dist_left_right_perc = left_right_perc[['term', 'left_bin', 'left_perc', 'right_bin', 'right_perc']]
        dist_left_right_perc.to_csv(word_dist_file.iloc[0], index=False)
        dist_left_right_heatmap = pd.DataFrame({'count' : dist_left_right_perc.groupby(['left_bin', 'right_bin'])['term'].count()}).reset_index()
        dist_left_right_heatmap.to_csv(heatmap_file.iloc[0], index=False)
        print("Iteration", i)
        print(word_dist_file.iloc[0])
        print(heatmap_file.iloc[0])
```
